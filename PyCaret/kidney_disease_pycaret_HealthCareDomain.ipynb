```python
import pandas as pd
import numpy as np


dataset=pd.read_csv('/content/kidney_dis_cleaneddata_pycarat.csv')
dataset = dataset.drop(['Unnamed: 0'],axis=1)
dataset
```

```python
dataset.columns
```

```python
#for training set
data=dataset.sample(frac=0.8,random_state=42).reset_index(drop=True)
#for test set(rename col for test after frac =0.8)
data_unseen=dataset.drop(data.index).reset_index(drop=True)
```

```python
print(f'training set: {data.shape}\ntest set: {data_unseen.shape}')
```

```python
# Install PyCaret directly from the main development branch on GitHub
#!pip install "git+https://github.com/pycaret/pycaret.git@master"
```

```python
from pycaret.classification import *
```

```python
classification_model=setup(data=data,target='classification',session_id=100)
```

```python
# comapre model performance

compare_models()
```

```python
# random forest  giving good accuracy create rf :

rf=create_model('rf')
```

```python
# hyperparameter tuning /(fine tuning):(original vs tuned--> which one is good?)

tuned_rf_Model=tune_model(rf)
```

```python
#evaluate the model -> original model is better already its 99.55% so will g for original model

evaluate_model(rf)
```

```python
# predict unseen data

unseen_prediction=predict_model(rf,data=data_unseen)
unseen_prediction.head()
```

```python
#classification is actual prediction-Original/ True Label
#prediction_label- model's final pred
#prediction_score- the Probability of the Prediction (the Confidence)
```
